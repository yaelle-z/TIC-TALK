# How to use this notebook:
# This notebook applies a shot classification model to a set of images.

# The main code is in cell 2.2.

# Input:
# - A shot classification model in ONNX format
# - A folder of images from a recorded performance (with subfolders)

# → Processing:
# The model is applied to each image to predict the shot type.

# Output:
# - A folder containing the sorted images,
#   with one subfolder per predicted class (named after the class code)


```python
## 2.2. Sorting images by shot type for a folder containing subfolders of images (entire corpus)
```

```python
import os
import shutil
from tqdm import tqdm
from ultralytics import YOLO

# path to the folder containing subfolders with the images to classify
input_folder = ''
# path to the folder where sorted images will be saved
output_folder = ''
# path to the YOLOv8 model in ONNX format
model_path = ''

# initialize the YOLOv8 model
model = YOLO(model_path)

# create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# loop through each subfolder in the input folder
for subfolder in os.listdir(input_folder):
    subfolder_path = os.path.join(input_folder, subfolder)
    
    # check if it is a folder
    if os.path.isdir(subfolder_path):
        # loop through all images in the subfolder
        for image_file in tqdm(os.listdir(subfolder_path), desc=f"Processing {subfolder}"):
            image_path = os.path.join(subfolder_path, image_file)
            
            # apply the classification model to the image
            results = model(image_path, imgsz=(224, 224))
            
            # get the predicted probabilities for each class
            probs = results[0].probs
            
            # get the index of the class with the highest probability
            top1_class_index = probs.top1
            
            # create a folder for the predicted class if it doesn't exist
            category_folder = os.path.join(output_folder, f"classe_{top1_class_index}")
            os.makedirs(category_folder, exist_ok=True)
            
            # copy the image to the corresponding output folder
            shutil.copy(image_path, category_folder)

print("Done!")
```
